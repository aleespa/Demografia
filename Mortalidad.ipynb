```python
from __future__ import division
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./conjunto_de_datos/defun15.tar.gz')
df.head()
```

```python
sorted(list(df.columns))
```

```python
df[df.lista_mex == "55"].head()
```

```python
df1 = df[(df.lista_mex == "55") ][df.ent_ocurr == 8][[ 'edad_agru','sexo']]
df1['TOTAL'] = 1
df1.head()
```

```python
df2 = df1.groupby(['edad_agru','sexo'], sort=False).count().reset_index()
df2.head()
sum(df2.TOTAL)
```

```python
gpe = pd.read_csv('./catalogos/deedadagrup.csv')
gpe.head()
```

```python
MO = pd.merge(df2, gpe, how='inner', left_on='edad_agru', right_on='CVE').sort_values(['CVE','sexo'])[['DESCRIP','sexo','TOTAL']]
MOH = MO[MO.sexo == 1][['DESCRIP', 'TOTAL']].reset_index()[['DESCRIP', 'TOTAL']]
MOM = MO[MO.sexo == 2] [['DESCRIP', 'TOTAL']].reset_index()[['DESCRIP', 'TOTAL']]
```

```python
MOH.to_csv('MOH.csv')
MOM.to_csv('MOM.csv')
```
